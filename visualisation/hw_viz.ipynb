```python
#функция года выпуска
def year_release_date(stats):
    stats['rel_year'] = stats.apply(lambda x: (str(x['release_date'])[0:4]), axis=1)
    
year_release_date(hw)
```

```python
#какой жанр самый популярный среди режиссеров (больше съемок)
hw5=hw[['genres','title']].groupby('genres').count().sort_values('title', ascending=False).reset_index().head(10)
hw5.head(10).style.bar()
#самый популярный жанра драма, на втором комедия
```

```python
#лист жанров
hw4=hw.dropna(subset=['genres'])
t=set()
f=()
for i in range(0, len(hw4)):
    f=set(hw4.iloc[i]['genres'].split())
    t=t.union(f)
print(list(t))
```

```python
#состаляем дата фрейм для проверки частоты кол-ва съемок отдельных жанров и их дохода
k={}
list_count=[]
list_revenue=[]
for i in list(t):
    d=0
    b=0
    for z in range(0, len(hw4)):
        f=list(hw4.iloc[z]['genres'].split())
        benefit=hw4.iloc[z]['revenue']
        if i in f:
            d+=1
            b+=benefit
    list_count.append(d)
    list_revenue.append(int(b/100000000))
k={'style':list(t),'count':list_count, 'benefit':list_revenue}
```

```python
data_p=pd.DataFrame(k)
```

```python
data_pp=data_p.groupby(['style']).sum().sort_values('count',ascending=False).head(15)
data_pp.plot(kind='barh')

plt.show()
#снимается больше всего драм, а доход получил жанр action, adventures
#комедий снимается меньше,а дохода больше чем от драм
#возможно была восстребованность в определенных годах в драме (всплеск просмотров ,высокий доход или рейтинг),повлекшие
#большое кол-во съемок в дальнейшем
```
